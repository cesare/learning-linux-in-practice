```python
import numpy as np
import pandas as pd
import matplotlib
import seaborn as sns
```

```python
df1 = pd.read_csv("results/1core-1process.tsv", delimiter="\t", header=None, names=("id", "elapsed", "progress"))
```

```python
sns.lmplot("elapsed", "id", data=df1, hue="id", markers=".")
```

```python
sns.lmplot("elapsed", "progress", data=df1, hue="id", scatter=True, markers=".")
```
